```
import numpy as np

class Variables:
  def __init__(self, data) -> None:
    self.data = data
    self.grad = None
    self.creator = None

  def set_creator(self, func):
    self.creator = func

  def backwardret(self):
    f = self.creator # 함수를 가져온다
    if f is not None:
      x = f.input # 함수의 입력을 가져온다
      x.grad = f.backwardret(self.grad) # 함수의 backward 메소드를 호출
      x.backwardret() # 하나 앞 변수의 backward 메소드를 호출(재귀)

  def backward(self):
    funcs = [self.creator]
    while funcs:
      f = funcs.pop()
      x,y = f.input, f.output
      x.grad = f.backward(y.grad)

      if x.creator is not None:
        funcs.append(x.creator)
```

```
class Function:
  def __call__(self, input):
    x = input.data
    y = self.forward(x)
    output = Variables(y)
    output.set_creator(self) # 출력 변수에 창조자를 설정
    self.input = input # 입력 변수를 기억 보관 이것은
    # 나중에 backward 메서드에서 함수에 입력한 변수가
    # 필요할 때 self.input에서 가져와 사용할 수 있다.
    self.output = output # 출력도 저장
    return output

  def forward(self, x):
    raise NotImplementedError()

  def backward(self, gy):
    raise NotImplementedError()
```

```
class Square(Function):
  def forward(self, x):
    y = x ** 2
    return y

  def backward(self, gy):
    x = self.input.data
    gx = 2 * x * gy
    return gx
```

```
class Exp(Function):
  def forward(self, x):
    y = np.exp(x)
    return y

  def backward(self, gy):
    x = self.input.data
    gx = np.exp(x) * gy
    return gx
```

```
A = Square()
B = Exp()
C = Square()

x = Variables(np.array(0.5))
a = A(x)
b = B(a)
y = C(b)

y.grad = np.array(1.0)
y.backward()
print(x.grad)
print("역전파")
print('--------------------')

b.grad = C.backward(y.grad)
a.grad = B.backward(b.grad)
x.grad = A.backward(a.grad)
print(x.grad)

# 계산 그래프의 노드들을 거꾸로 거슬러 올라간다.
assert y.creator == C
assert y.creator.input == b
assert y.creator.input.creator == B
assert y.creator.input.creator.input == a
assert y.creator.input.creator.input.creator == A
assert y.creator.input.creator.input.creator.input == x
```

```
x = 2
w = 3
b = 1
yT = 10
lr = 0.01

for epoch in range (200):
  y = x * w + 1 * b
  E = (y - yT) * (y - yT) / 2
  yE = y - yT
  wE = yE * x
  bE = yE * 1
  w -= lr * wE
  b -= lr * bE
  print("epoch = " + str(epoch + 0))
  print(" y : " + str(y))
  print(" w : " + str(w))
  print(" b ; " + str(b))

  if E < 0.0000001:
    break
```

```
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim

x1, x2 = 2, 3
t1, t2 = 27, -30
w1, w3 = 3, 5
w2, w4 = 4, 6
b1, b2 = 1, 2

for epoch in range(2000):
  print('epoch = %d' %epoch)
  y1 = x1*w1 + x2*w2 + 1*b1
  y2 = x1*w3 + x2*w4 + 1*b2
  print('y1, y2 = %6.3f, %6.3f' %(y1, y2))

  E = (y1-t1)**2/2 + (y2-t2)**2/2
  print('E = %.7f' %E)
  if E < 0.0000001:
    break

  y1b, y2b = y1 - t1, y2 - t2
  x1b, x2b = y1b*w1 + y2b*w3, y1b*w2+y2b*w4
  w1b, w3b = x1*y1b, x1*y2b
  w2b, w4b = x2*y1b, x2*y2b
  b1b, b2b = 1*y1b, 1*y2b
  print(' x1b, x2b = %6.3f, %6.3f' %(x1b, x2b))
  print(' w1b, w3b = %6.3f, %6.3f' %(w1b, w3b))
  print(' w2b, w4b = %6.3f, %6.3f' %(w2b, w4b))
  print(' b1b, b2b = %6.3f, %6.3f' %(b1b, b2b))

  lr = 0.01
  w1, w3 = w1 - lr*w1b, w3 - lr*w3b
  w2, w4 = w2 - lr*w2b, w4 - lr*w4b
  b1, b2 = b1 - lr*b1b, b2 - lr*b2b
```

```
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim

X = torch.FloatTensor([[2,3]])  # 입력 데이터
YT = torch.FloatTensor([[27]])  # 목표데이터(라벨)
W = torch.FloatTensor([[3,4]])  # 가중치
B = torch.FloatTensor([[1]])  # 편향
model = nn.Sequential(
    nn.Linear(2,1)  # 입력층, 출력층 노드의 갯수를 각각 1개로 설정한다
) # 신경망 모양 결정 (W, B 내부적 준비)
print(model)  # 신경망 모양 출력

with torch.no_grad():
  model[0].weight = nn.Parameter(W)
  model[0].bias = nn.Parameter(B)

loss_fn = nn.MSELoss()  # 2공식 오차 계산 함수
optimizer = optim.SGD(model.parameters(), lr=0.01)

for epoch in range(1000):
  Y = model(X)        # 1공식, 순전파
  E = loss_fn(Y, YT)  # 2공식 , 오차계산
  optimizer.zero_grad()
  E.backward()        # 6공식, 오차역전파
  optimizer.step()    # 7공식, 학습

  if epoch%100 == 99:
    print(epoch, E.item())

print(f'W = {model[0].weight.data}')
print(f'B = {model[0].bias.data}')

Y = model(X)
print(Y.data)
```

# 퍼셉트론과 인간의 할 일
퍼셉트론의 매개변수 값을 정하는 것은 컴퓨터가 아니라 우리 인간입니다.
인간이 직접 진리표라는 '학습 데이터'를 보면서 매개변수의 값을 생각했습니다.
기계학습 문제는 이 매개변수의 값을 정하는 작업을 컴퓨터가 자동으로 하도록 합니다. 학습이란 적절한 매개변수 값을 정하는 작업이며, 사람은 퍼셉트론의 구조(모델)을 고민하고 컴퓨터에 학습할 데이터를 주는 일을 합니다.

```
# 간단한 퍼셉트론 구현
def AND(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  tmp = x1*w1 + x2*w2
  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1

print(AND(0, 0))
print(AND(1, 0))
print(AND(0, 1))
print(AND(1, 1))
```

# 가중치와 편향의 도입
앞서 수식에서 theta 대신에 -b로 치환하면 퍼셉트론의 동작이 다음 식2.2처럼 됩니다.


*(식2.2)*  **  y = 0 (b + w1x1 + w2x2 <= 0), y = 1 (b + w1x1 + w2x2 > 0)**
앞의 수식과 식2.2는 기호만 바뀌었을 뿐, 그 의미는 같습니다.
여기서 b를 **편향**bias 라고 하며 w1과 w2는 그대로 가중치weight입니다.


```
import numpy as np
x = np.array([0, 1])  # 입력
w = np.array([0.5, 0.5])  # 가중치
b = -0.7  # 편향
print(w*x)
print(np.sum(w*x))
print(np.sum(w*x) + b)
```

```
def NAND(x1, x2):
  x = np.array([x1, x2])
  w = np.array([-0.5, -0.5])
  b = 0.7
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  elif tmp > 0:
    return 1

def OR(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.2
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1

print(AND(1, 0))
print(NAND(1, 0))
print(NAND(1, 1))
print(OR(0, 1))
print(OR(1, 1))
```

# 퍼셉트론의 한계
x1과 x2 중 한쪽이 1일 때만 1을 출력하는 배타적 논리합 XOR 게이트를 퍼셉트론으로 구현하려면 가중치 매개변수 값을 어떻게 설정하면 될까요?
일반 단층 퍼셉트론 만으로는 표현할 수 없으며 다층 퍼셉트론으로 해결 할 수 있습니다.
앞에서 정의한 AND,OR,NAND 게이트들을 적당히 조합하면 XOR 게이트와 같은 출력을 만들 수 있습니다.
그러면 XOR 게이트를 파이썬 코드로 구현해 봅시다



```
def XOR(x1, x2):     # 0층에 해당하는 노드
  s1 = NAND(x1, x2)  # 1층에 해당하는 노드
  s2 = OR(x1, x2)    # 1층에 해당하는 노드
  y = AND(s1, s2)    # 2층에 해당하는 노드
  return y

print(XOR(0, 0))
print(XOR(1, 0))
print(XOR(0, 1))
print(XOR(1, 1))
```

이상으로 2층 구조를 사용해 퍼셉트론으로 XOR 게이트를 구현 할 수 있게 되었습니다.
다시 말해 단층 퍼셉트론으로는 표현하지 못한 것을 층을 하나 늘려서 구현할 수 있었습니다.
이처럼 퍼셉트론은 층을 쌓아(깊게 하여) 더 다양한 것을 표현수 있습니다.
컴퓨터도 다층 퍼셉트론으로 만들 수 있습니다.
정확히는 비선형인 시그모이드 함수를 활성화 함수로 이용하면 임의의 함수를 표현할 수 있다는 사실이 증명되었습니다.(3장 참조)

단층 퍼셉트론은 직선형 영역만 표현할 수 있고, 다층 퍼셉트론은 비선형 영역도 표현할 수 있다.

# 신경망
퍼셉트론은 복잡한 함수도 표현할 수 있다는 장점이 있지만, 원하는 결과를 출력하도록 가중치 값을 적절히 정하는 작업은 여전히 사람이 수동으로 한다는 단점도 갖고있다는 것이다.
신경망은 이러한 퍼셉트론의 단점을 해결해줍니다.
무슨 말인고 하니, 가중치 매개변수의 적절한 값을 데이터로부터 자동으로 학습하는 능력이 이제부터 살펴볼 신경망의 중요한 성질입니다.

**활성화 함수**
입력 신호의 총합을 출력 신호로 변환하는 함수

y = 0 : (b + w1x1 + w2x2 <= 0)
y = 1 : (b + w1x1 + w2x2 > 0)

상기 식을 다음과 같이 표현할 수 있다.

y = h(b + w1x1 + w2x2) --- [식3.2]
h(x) = 0 (x <= 0)
h(x) = 1 (x > 0)

여기서 h(x)라는 함수를 활성화 함수라고 합니다.
즉 입력 신호의 총합이 활성화를 일으키는지를 정하는 역할을 합니다.
[식3.2]는 가중치가 곱해진 입력 신호의 총합을 계산하고, 그 합을 활성화 함수에 입력해 결과를 내는 2단계로 처리됩니다.
그래서 이 식은 다음과 같은 2단계로 나눌 수 있습니다.

          a = b + w1x1 =+ w2x2  [식3.4]
              y = h(a)  [식3.5]

활성화 함수가 퍼셉트론에서 신경망으로 가기 위한 길잡이랍니다.
